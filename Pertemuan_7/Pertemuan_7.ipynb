### 1. Lanskap Machine Learning
Secara garis besar, algoritma *Machine Learning* terbagi menjadi beberapa paradigma utama, di antaranya:
* **Supervised Learning:** Model belajar dari pasangan data input ($X$) dan label target ($y$) yang sudah diketahui kunci jawabannya. Paradigma ini dibagi lagi menjadi dua tugas utama, yaitu **Klasifikasi** untuk menghasilkan output berupa kelas/label diskret dan **Regresi** untuk memprediksi nilai kontinu.
* **Unsupervised Learning:** Model mencari pola tersembunyi dari data tanpa label karena tidak ada $y$ yang diketahui. Tugas utamanya meliputi **Clustering** (mengelompokkan data yang mirip) dan **Dimensionality Reduction** (reduksi fitur sambil mempertahankan informasi).

### 2. Konsep dan Mekanisme Regresi Linear
Regresi Linear adalah algoritma yang mencari garis lurus terbaik yang mendeskripsikan hubungan antara variabel input ($X$) dan variabel output ($y$) yang bersifat kontinu. Persamaan matematis untuk Regresi Linear Sederhana dinyatakan sebagai:

$$\hat{y} = \beta_{0} + \beta_{1} \cdot x$$

Di mana $\hat{y}$ adalah nilai yang diprediksi, $\beta_{0}$ merupakan *intercept* (nilai $y$ saat $x=0$), $\beta_{1}$ adalah *slope* atau koefisien kemiringan yang menunjukkan seberapa banyak $y$ berubah per unit $x$, dan $x$ adalah variabel input atau fitur. Algoritma ini disebut *Multiple Linear Regression* ketika melibatkan banyak variabel input untuk memprediksi satu output.

Untuk menemukan garis terbaik, model harus meminimalkan nilai kesalahan melalui **Cost Function (Fungsi Biaya)**, yang diukur menggunakan *Mean Squared Error* (MSE). MSE menghitung rata-rata dari selisih kuadrat antara nilai aktual ($y_{i}$) dan nilai prediksi ($\hat{y}_{i}$). Proses pencarian nilai koefisien $\beta$ yang optimal dilakukan menggunakan algoritma **Gradient Descent**, yaitu metode yang mengubah parameter secara bertahap dengan arah menuruni lembah fungsi biaya terdalam berdasarkan kontrol dari *learning rate* ($\alpha$).

### 3. Metrik Evaluasi Regresi
Untuk menguji keakuratan model pada data pengujian, terdapat empat metrik evaluasi utama yang digunakan:
* **MAE (Mean Absolute Error):** Menghitung rata-rata nilai absolut dari *error*. Metrik ini mudah diinterpretasikan karena satuannya sama dengan $y$ dan bersifat *robust* terhadap *outlier* karena memberikan penalti secara linier.
* **MSE (Mean Squared Error):** Menghitung rata-rata selisih kuadrat. Berguna untuk fungsi kerugian (*loss function*) internal saat melatih model karena sifatnya yang dapat diturunkan (*differentiable*), namun sulit diinterpretasikan karena satuannya ikut dikuadratkan.
* **RMSE (Root Mean Squared Error):** Akar kuadrat dari MSE. Memiliki satuan yang sama dengan $y$ dan memberikan penalti yang lebih besar untuk *outlier* dibanding MAE, menjadikannya metrik yang paling umum digunakan untuk *benchmarking*.
