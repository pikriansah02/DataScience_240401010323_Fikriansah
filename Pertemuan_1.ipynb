Ringkasan Materi Pertemuan 1 - Pengenalan Data Science
Pada pertemuan pertama, saya mempelajari konsep dasar Data Science sebagai bidang multidisiplin yang menggabungkan statistika, ilmu komputer, dan pengetahuan domain untuk menghasilkan informasi yang bermanfaat dari data. Data Science digunakan untuk membantu pengambilan keputusan berdasarkan data melalui proses analisis dan pemodelan.

Terdapat tiga pilar utama dalam Data Science, yaitu Statistika dan Matematika, Ilmu Komputer, serta Domain Knowledge. Ketiga pilar tersebut saling melengkapi dalam proses pengolahan dan analisis data.

Saya juga mempelajari fenomena Big Data yang terdiri dari Volume, Velocity, Variety, dan Veracity. Selain itu, dikenalkan berbagai profesi di bidang Data Science seperti Data Analyst, Data Scientist, Machine Learning Engineer, dan Data Engineer beserta tugas utamanya.

Dalam pengelolaan proyek Data Science digunakan metodologi CRISP-DM yang terdiri dari enam tahapan, yaitu Business Understanding, Data Understanding, Data Preparation, Modeling, Evaluation, dan Deployment. Metodologi ini bersifat iteratif sehingga memungkinkan perbaikan pada setiap tahap.

Pada sesi praktikum, saya mempelajari penggunaan Google Colab, GitHub, dan dasar-dasar Python seperti perintah print(), variabel, tipe data, list, perulangan for, serta pembuatan fungsi sederhana. Tools yang digunakan dalam pembelajaran ini adalah Python, Google Colab, Jupyter Notebook, dan GitHub.
