## Ringkasan Materi

Pada Sesi 4, saya mempelajari konsep statistika dasar dan analisis data yang merupakan komponen fundamental dalam eksplorasi data (*Exploratory Data Analysis* atau EDA). Materi pembelajaran dibagi menjadi tiga bagian utama, yaitu statistika deskriptif, analisis distribusi univariat, serta analisis bivariat menggunakan bahasa pemrograman Python.

Bagian pertama membahas **Statistika Deskriptif** yang mencakup ukuran pemusatan dan penyebaran data. Ukuran pemusatan data yang dipelajari meliputi *Mean* (rata-rata), *Median* (nilai tengah), dan *Modus* (nilai dengan frekuensi tertinggi). Pemilihan ketiga ukuran ini sangat bergantung pada karakteristik data: *mean* digunakan untuk data numerik dengan distribusi simetris namun sensitif terhadap *outlier*; *median* lebih bersifat *robust* (tahan) terhadap *outlier* sehingga cocok untuk data yang miring (*skewed*); sedangkan *modus* digunakan untuk data kategorik. Selain itu, saya juga mempelajari varians dan standar deviasi sebagai pengukur sebaran data dari nilai rata-ratanya, serta kuartil, persentil, dan *Interquartile Range* (IQR) yang digunakan untuk memotong data menjadi beberapa bagian sekaligus menentukan batas pencilan (*outlier*) menggunakan metode IQR Fence.

Bagian kedua berfokus pada **Distribusi & Analisis Univariat** (analisis satu variabel) untuk mengidentifikasi bentuk visual dari sebaran data. Bentuk distribusi data dapat dikategorikan menjadi distribusi Normal (simetris di mana nilai *mean* = *median* = *modus*), *Right-Skewed* (miring kanan, nilai *mean* > *median*), *Left-Skewed* (miring kiri, nilai *mean* < *median*), dan *Bimodal* (memiliki dua puncak frekuensi yang mengindikasikan adanya dua sub-populasi). Karakteristik bentuk ini diukur secara matematis melalui nilai *Skewness* (kemiringan) dan *Kurtosis* (ketebalan ekor/keruncingan grafik). Secara praktis, analisis univariat ini diimplementasikan di Python menggunakan visualisasi **Histogram** dan **KDE** (*Kernel Density Estimation*) untuk melihat kepadatan data, serta **Boxplot** dan **Violin Plot** untuk membandingkan ringkasan lima angka statistika dan bentuk distribusi antar-grup secara lebih informatif.

Bagian ketiga mengulas tentang **Analisis Bivariat** yang digunakan untuk menyelidiki hubungan antara dua variabel numerik. Visualisasi utama yang digunakan adalah **Scatter Plot** yang mampu mendeteksi pola korelasi positif (kedua variabel naik bersamaan), korelasi negatif (satu variabel naik, variabel lain turun), tidak berkorelasi (titik tersebar acak), maupun hubungan non-linear. Kekuatan hubungan antar-variabel ini diukur menggunakan nilai koefisien korelasi dengan rentang antara -1 hingga +1. **Korelasi Pearson (r)** dipilih jika data bertipe numerik, berdistribusi normal, dan memiliki hubungan linear, sedangkan **Korelasi Spearman ($\rho$)** digunakan apabila data bertipe ordinal, tidak berdistribusi normal, atau mengandung *outlier*. Agar matriks korelasi dari banyak variabel lebih mudah diinterpretasikan, data divisualisasikan menggunakan **Heatmap Korelasi** dengan gradasi warna yang menunjukkan kekuatan hubungan positif atau negatif secara instan.
