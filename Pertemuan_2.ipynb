## Ringkasan Materi

Pada pertemuan kedua, saya mempelajari struktur data bawaan Python yang terdiri dari List, Tuple, Dictionary, dan Set. Setiap struktur data memiliki karakteristik dan kegunaan yang berbeda. List digunakan untuk menyimpan data yang dapat diubah, Tuple digunakan untuk data yang bersifat tetap, Dictionary digunakan untuk menyimpan pasangan key-value, sedangkan Set digunakan untuk menyimpan data unik tanpa duplikasi.

Selain struktur data Python, saya juga mempelajari library NumPy yang digunakan untuk komputasi numerik. NumPy menyediakan objek ndarray yang memungkinkan pengolahan data numerik secara lebih cepat dan efisien dibandingkan list Python biasa. Saya mempelajari cara membuat array, memahami atribut array, serta melakukan operasi matematika dan statistik menggunakan konsep vektorisasi.

Materi berikutnya membahas Pandas sebagai library utama dalam manipulasi dan analisis data. Saya mempelajari penggunaan Series dan DataFrame sebagai struktur data utama dalam Pandas. Selain itu, saya belajar membaca data dari file CSV menggunakan fungsi read_csv(), melakukan eksplorasi data, filtering dan seleksi data menggunakan boolean indexing, loc(), dan iloc(), serta melakukan agregasi data menggunakan groupby().

Pada sesi praktikum, saya menggunakan dataset Titanic untuk melakukan analisis data sederhana, seperti menghitung jumlah data, memeriksa missing values, menghitung tingkat keselamatan penumpang, melakukan filtering berdasarkan kategori tertentu, serta membuat analisis menggunakan operasi groupby(). Materi ini memberikan dasar yang kuat untuk proses pengolahan dan analisis data pada tahap selanjutnya.
