## Ringkasan Materi

Pada Sesi 6, saya mempelajari tahap persiapan data (*data preparation/preprocessing*), sebuah fase krusial sebelum data dimasukkan ke dalam algoritma *Machine Learning*. Karena algoritma *Machine Learning* pada dasarnya bekerja menggunakan operasi matematika (seperti perkalian matriks dan kalkulasi jarak), semua input data harus disajikan dalam format numerik yang seragam. Kegagalan dalam menyiapkan data dengan benar akan berakibat pada kegagalan model secara keseluruhan sesuai prinsip *Garbage In, Garbage Out*. Pembelajaran kali ini berfokus pada tiga teknik utama: *encoding* data kategorikal, *feature scaling*, dan pembagian dataset (*train-test split*).

### 1. Pipeline Persiapan Data & Data Leakage
Urutan langkah dalam mempersiapkan data sangat menentukan validitas performa model. Urutan terstruktur yang direkomendasikan adalah melakukan Analisis Eksplorasi Data (EDA), melakukan *encoding* kategorikal, menangani *missing values*, membagi dataset (*Train-Test Split*), menjalankan *Feature Scaling*, dan diakhiri dengan pelatihan model.

Proses *Train-Test Split* wajib dilakukan **sebelum** proses pembentukan skala (*fitting scaler/imputer*). Jika pemenuhan skala (*fit*) diaplikasikan pada seluruh data sekaligus, informasi dari *test set* akan bocor ke dalam proses pelatihan model. Fenomena anomali ini disebut sebagai **Data Leakage**, yang mengakibatkan metrik evaluasi model menjadi terlalu optimis dan tidak mencerminkan performa asli di data nyata.

### 2. Encoding Data Kategorikal
Data kategorikal merepresentasikan kelompok atau kategori, bukan nilai numerik yang bermakna secara matematis sehingga harus dikonversi menjadi representasi angka. Metode transformasinya terbagi menjadi tiga jenis:
* **Label Encoding:** Menetapkan satu bilangan bulat unik untuk setiap kategori secara otomatis berdasarkan urutan alfabet. Metode ini hanya cocok digunakan untuk kolom target (*y*) biner/multi-kelas, atribut biner, atau fitur ordinal. Metode ini harus dihindari pada input fitur nominal agar model tidak salah mengasumsikan adanya urutan matematis.
* **One-Hot Encoding (OHE):** Menciptakan satu kolom biner baru (0 dan 1) untuk setiap nilai unik dalam kolom kategorikal. Guna menghindari **Dummy Variable Trap** (multi-kolinearitas sempurna), salah satu kolom biner harus dihapus menggunakan parameter `drop_first=True` atau `drop='first'`. Teknik OHE sebaiknya dihindari apabila suatu kolom memiliki kardinalitas tinggi (*high cardinality* > 100 kategori) karena akan membengkakkan dimensi dataset.
* **Ordinal Encoding:** Digunakan khusus untuk data kategorikal yang memiliki urutan alami (seperti tingkat pendidikan). Penomoran angka ditentukan secara eksplisit berdasarkan urutan makna domain, bukan berdasarkan abjad.

### 3. Scaling & Normalisasi Fitur
*Feature scaling* menyeragamkan rentang nilai fitur numerik agar fitur dengan skala besar tidak mendominasi perhitungan jarak algoritma terhadap fitur berskala kecil. Terdapat tiga jenis alat penskalaan di library *scikit-learn*:
* **MinMaxScaler:** Mentransformasikan nilai data ke dalam rentang tetap `[0, 1]`. Teknik ini banyak digunakan pada arsitektur *Neural Network* dan pengolahan citra, namun memiliki kelemahan karena sangat sensitif terhadap gangguan data pencilan (*outlier* ekstrem).
* **StandardScaler:** Mengubah distribusi data menjadi terstandarisasi dengan nilai rata-rata (*mean*) = 0 dan standar deviasi = 1. Pendekatan ini tidak menjamin rentang output tetap, namun membuatnya lebih *robust* (toleran) terhadap pencilan sedang dibandingkan MinMaxScaler.
* **RobustScaler:** Menggunakan nilai *median* dan *Interquartile Range* (IQR) untuk penskalaan. Pilihan ini sangat optimal jika dataset memiliki banyak pencilan ekstrem yang tidak bisa dihapus dari data.

### 4. Train-Test Split & Stratifikasi
Pembagian dataset menjadi *training set* (untuk belajar) dan *test set* (untuk evaluasi) bertujuan untuk menguji kemampuan generalisasi model terhadap data baru, sekaligus mengidentifikasi gejala *overfitting*. Proporsi pembagian yang umum berkisar pada rasio 70:30 untuk data kecil (< 1.000 baris), 80:20 untuk ukuran data medium, atau 90:10 untuk kumpulan data berskala besar.

Pada kasus klasifikasi data dengan kelas target yang tidak seimbang (*imbalanced class*), pengacakan acak biasa berisiko membagi proporsi kelas secara timpang. Solusinya adalah menyertakan parameter `stratify` agar persentase proporsi masing-masing kelas target tetap terjaga secara seimbang, baik di dalam data *train* maupun data *test*.
