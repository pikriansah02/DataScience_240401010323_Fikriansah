## Ringkasan Materi

Pada Sesi 5, saya mempelajari prinsip-prinsip visualisasi data yang efektif, pemanfaatan dua library utama Python (Matplotlib dan Seaborn), serta kerangka kerja sistematis untuk membaca grafik guna menarik sebuah insight. Visualisasi data berperan krusial sebagai jembatan antara angka mentah dan pemahaman manusia, baik untuk kebutuhan eksplorasi (EDA) maupun komunikasi penemuan kepada pemangku kepentingan.

Terdapat lima prinsip utama untuk menghasilkan visualisasi yang efektif:
1. **Kejelasan (Clarity):** Setiap elemen visual harus memiliki tujuan jelas dan menghindari dekorasi tanpa makna (chartjunk) demi memaksimalkan data-ink ratio.
2. **Akurasi (Accuracy):** Grafik tidak boleh memanipulasi fakta, seperti keharusan sumbu Y dimulai dari nol pada bar chart agar perbedaan nilai tidak membesar secara artifisial.
3. **Efisiensi (Efficiency):** Menampilkan informasi sebanyak mungkin dengan elemen visual sesedikit mungkin untuk meminimalkan beban kognitif pembaca.
4. **Estetika (Aesthetics):** Mengedepankan konsistensi jenis huruf (font), ukuran, spasi, serta penggunaan palet warna yang bermakna dan ramah buta warna.
5. **Konteks (Context):** Menyediakan judul deskriptif yang spesifik, label sumbu yang lengkap dengan satuan, serta legenda yang jelas agar grafik dapat dipahami tanpa penjelasan tambahan.

Dalam penerapannya, pemilihan jenis grafik disesuaikan dengan tujuan komunikasi, seperti bar chart untuk perbandingan kategori, line chart untuk tren waktu, histogram/KDE untuk distribusi univariat, dan scatter plot untuk melihat hubungan dua variabel.

Pada bagian pemrograman, Matplotlib dipelajari sebagai library dasar berspesifikasi low-level control yang menggunakan hierarki objek Figure (kanvas utama), Axes (area plot/panel), dan Axis (sumbu koordinat). Sementara Seaborn merupakan library tingkat tinggi (high-level) yang dibangun di atas Matplotlib untuk menyederhanakan pembuatan grafik statistik secara ringkas dan terintegrasi langsung dengan Pandas DataFrame.

Materi terakhir membahas tentang framework analisis "What? So what? Now what?" untuk membaca grafik secara sistematis:
* **What?:** Deskripsi faktual dan objektif mengenai apa yang diukur, rentang nilai, dan satuan pada grafik.
* **So what?:** Interpretasi untuk membedah pola, tren, nilai ekstrem, atau anomali guna menghasilkan sebuah insight bermakna.
* **Now what?:** Menentukan implikasi praktis berupa rekomendasi tindakan nyata atau agenda eksplorasi lanjutan berbasis data.
