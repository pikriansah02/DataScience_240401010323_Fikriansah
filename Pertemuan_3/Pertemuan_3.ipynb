## Ringkasan Materi

Pada pertemuan ketiga, saya mempelajari salah satu tahapan paling kritis dalam proyek Data Science, yaitu pembersihan data (data cleaning). Tahap ini sangat krusial karena kualitas analisis dan model machine learning sangat dibatasi oleh kualitas data yang digunakan (prinsip 'Garbage In, Garbage Out'). Survei industri bahkan menunjukkan bahwa aktivitas persiapan dan pembersihan data menyita sekitar 37.75% waktu kerja seorang Data Scientist sehari-hari.

Secara umum, terdapat enam kategori masalah kualitas data yang sering ditemui, yaitu missing values, duplikat, inkonsistensi format, outlier, nilai tidak valid, dan skala yang tidak seragam. Pada materi missing values, saya mempelajari tiga mekanisme hilangnya data, yaitu MCAR (Missing Completely At Random), MAR (Missing At Random), dan MNAR (Missing Not At Random). Untuk mendeteksinya, kita dapat memanfaatkan fungsi `isnull().sum()` serta visualisasi pola menggunakan library `missingno` seperti `msno.matrix()` dan `msno.heatmap()`. Strategi penanganannya meliputi penghapusan data (dropping) jika data bersifat MCAR dan jumlahnya relatif kecil, imputasi statistik menggunakan mean, median, atau modus, forward/backward fill untuk data time-series, interpolasi linear, hingga imputasi berbasis model menggunakan KNN Imputer.

Selain missing values, modul ini membahas cara mendeteksi dan menangani data duplikat menggunakan `duplicated()` and `drop_duplicates()`. Kami juga mempelajari penanganan outlier (nilai ekstrem) melalui dua pendekatan utama, yaitu metode IQR Fence (Pagar Kuartil) yang robust, dan Metode Z-Score yang efektif untuk data berdistribusi normal. Outlier yang terdeteksi dapat ditangani dengan cara dihapus atau disesuaikan nilainya menggunakan teknik Winsorization (Capping) melalui fungsi `.clip()`. Materi terakhir membahas tentang ekstraksi data dari format JSON dan REST API. Pandas menyediakan fungsi `pd.read_json()` and `json_normalize()` untuk membaca dan meratakan struktur data JSON yang bersarang (nested). Sementara untuk mengakses REST API publik, kami menggunakan library `requests` dengan metode GET. Modul ini juga menekankan beberapa best practices dalam mengakses API, seperti memeriksa status_code, menerapkan timeout, menangani exception koneksi, memperhatikan rate limiting, dan menjaga keamanan API key.

## Kesimpulan

Pada pertemuan ini, saya memahami bahwa proses data cleaning adalah fondasi utama untuk menghasilkan analisis data yang valid dan model prediksi yang akurat. Saya telah mempelajari taksonomi masalah kualitas data serta berbagai metodologi terstruktur untuk menangani missing values, data duplikat, dan outlier menggunakan ekosistem Python (Pandas, NumPy, Scipy, dan Scikit-Learn). Selain itu, saya juga memahami cara memperluas sumber perolehan data secara modern melalui pemrosesan file JSON lokal maupun web service berbasis REST API.

Proses ini mencakup pembersihan duplikat, normalisasi string, imputasi nilai kosong dengan median dan modus, hingga pembatasan outlier menggunakan teknik IQR capping, yang diakhiri dengan proses validasi menggunakan fungsi `assert`.

**Temuan Utama:** Langkah pembersihan data tidak boleh dilakukan secara sembarangan. Pemilihan metode imputasi missing values (seperti penggunaan median untuk distribusi data yang skewed) serta penanganan outlier yang tepat sangat memengaruhi distribusi asli data dan performa model kedepannya.
